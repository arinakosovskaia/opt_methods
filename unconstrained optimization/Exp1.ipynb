```python
import pandas as pd
import numpy as np
import scipy.optimize
import matplotlib.pyplot as plt

from oracles import QuadraticOracle
from optimization import gradient_descent
import plot_trajectory_2d
from plot_trajectory_2d import plot_levels, plot_trajectory
```

```python
np.random.seed(0)
```

```python
A = [np.array([ [1, 0], [0, 1]]), 
     np.array([ [1, 0], [0, 10]]), 
     np.array([ [1, 2], [2, 5]])]
```

```python
B = [np.array([ 0, 0 ]), 
     np.array([ 0, 0 ]),
     np.array([ 0, 0 ]) ]
```

```python
#1й эксперимент
i=0
x0 = [-4, 1]
for a, b in zip(A, B):
  k = np.linalg.cond(a)
  oracle = QuadraticOracle(a, b)
  i += 1
  history = gradient_descent(oracle, x0, trace=True)[2]
  plt.figure()
  plot_levels(oracle.func)
  plot_trajectory_2d.plot_trajectory(oracle.func, history['x'])
  plt.title(str('k: ' + str(k) + ', iterations: ' + str(len(history['x'])-1)))
  name = str(i) + '.png'
  plt.savefig(name, bbox_inches="tight")
```

```python
#3й эксперимент
i=13
x0 = [3, 5]
for a, b in zip(A, B):
  k = np.linalg.cond(a)
  oracle = QuadraticOracle(a, b)
  for method in ['Wolfe', 'Armijo', 'Constant']:
    i += 1
    if method != 'Constant':
      history = gradient_descent(oracle, x0, line_search_options={'method': method}, trace=True)[2]
    else:
      history = gradient_descent(oracle, x0, line_search_options={'method': method, 'c': 0.1}, trace=True)[2]
    plt.figure()
    plot_levels(oracle.func)
    plot_trajectory_2d.plot_trajectory(oracle.func, history['x'])
    plt.title(str('k: ' + str(k) + ', method: ' + str(method) + ', iterations: ' + str(len(history['x'])-1)))
    name = str(i) + '.png'
    plt.savefig(name, bbox_inches="tight")
```
